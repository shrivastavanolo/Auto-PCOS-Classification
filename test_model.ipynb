```python
from keras.models import load_model
import numpy as np
```

```python
#import the saved checkpoint using load_model function
model = load_model(r"C:\Users\shrey\OneDrive\Documents\misahub\bestmodel.h5")
```

```python
from keras.preprocessing import image
import tensorflow as tf

#function for predicting health using image path. 1 means healthy and 0 means not healthy
def predictimage(path):
    
      img = tf.keras.utils.load_img(path,target_size=(224,224))
      i = tf.keras.utils.img_to_array(img)/255
      input_arr= np.array([i])
      input_arr.shape

      pred =float(model.predict(input_arr))
      if pred > 0.1:
            return 1
      else:
            return 0
```

```python
import os
import pandas as pd
```

```python
test_data = r"C:\Users\shrey\OneDrive\Documents\misahub\PCOSGen-test\images"

#create new dataframe for saving result
df1=pd.DataFrame(columns=["S. No.","Image Path","Healthy"])

#predict all images from test folder and save result to df1['Healthy']
for i,file in enumerate(os.listdir(test_data),start=1):
    file_path=os.path.join(test_data,file)
    pred=predictimage(file_path)
    df1.loc[len(df1.index)] = [i,file,pred]
```

```python
df1.to_excel("Testing_data_Result.xlsx")
```
